2.  What is an interpreter in Python?
 - An interpreter in Python is a program that executes the code line by line, converting Python code into machine-readable instructions instantly.

3.  What are pre-defined keywords in Python?
 - Pre-defined keywords are special reserved words in Python that already have a fixed meaning and cannot be used as variable names.
Examples: if, else, for, while, def, return, True, False.

4. Can keywords be used as variable names?
 - No, keywords cannot be used as variable names because they are reserved for Python’s own syntax and functions

5. What is mutability in Python?
 - Mutability means whether an object can be changed after it is created.

   Mutable → can change

     Immutable → cannot change

6. Why are lists mutable, but tuples are immutable?
 - Lists are meant for changing data

   Lists are designed for cases where you add, remove, or update items.

   So Python stores them in a flexible way in memory.

   You can do: append(), pop(), remove(), list[0] = 10 etc.

- Tuples are meant for fixed data

  Tuples are created to store data that should NOT change.

  Python makes them read-only to keep them safe.

  No append(), no pop(), no item assignment.
- One line summary

  Lists change because they’re designed for dynamic data.

  Tuples don’t change because they’re designed for fixed, safe, and faster data storage.
7.  What is the difference between “==” and “is” operators in Python?
- == → checks VALUE

  It checks whether the contents / values of two variables are the same.
-   is → checks IDENTITY (same memory location?)

    It checks whether both variables refer to the exact same object in memory.
8.  What are logical operators in Python?
- and

  Returns True only if both conditions are True.   
- or

  Returns True if any one of the conditions is True.
- not

  Reverses the result (True → False, False → True).  

9. What is type casting in Python?
- It is the process of changing a value’s data type, like converting:

  string → int

  int → float

  float → string

  list → tuple, etc.

 10. What is the difference between implicit and explicit type casting?
 - Implicit Type Casting (Automatic)

   Python automatically converts one type to another.

   Happens when there is no data loss.

   Usually during arithmetic operations.
- Explicit Type Casting (Manual)

  You convert the type yourself by using functions like:

  int()

  float()

  str()

  list(), tuple(), etc.

11. What is the purpose of conditional statements in Python?
- To control the flow of the program

  They allow the program to choose different paths based on conditions.

- To execute code only when a specific condition is True

  Instead of running everything blindly, Python checks:

  Is the condition True? → run the block

  Is it False? → skip it

-  To make decisions based on user input or data

   Like checking age, comparing values, validating login, etc.

- To add logic and intelligence to the program

  Without conditions, the program can’t think or react.

12. How does the elif statement work ?
- elif basically means "else if".
  It is used when you have multiple conditions, and you want Python to check them one by one.

- How it works step-by-step:

  Python checks the if condition first.

  If if is True, it runs that block and skips all elif and else.

  If if is False, Python goes to the first elif.

  If that elif is True, it runs the block and stops.

  If not, it checks the next elif.

  If all conditions are False, then the else runs (optional).

13. What is the difference between for and while loops?
- FOR loop

  Used when you know how many times the loop should run.

  Or when you want to loop through a list, string, or range.
- WHILE loop

  Used when you don’t know how many times the loop will run.

  It keeps running as long as the condition is True.

14. Describe a scenario where a while loop is more suitable than a for loop.
- Scenario Where a while Loop Is Better

  Example Situation: Keep asking the user for a correct password

  You don’t know how many attempts the user will take.
  It could be 1 time… 5 times… or 20 times.

  In this case, a while loop is perfect because it keeps running until the user enters the right password.

```python
# Write a Python program to convert user input from string to integer, float, and boolean types
print("1. Convert to Integer")
print("2. Convert to Float")
print("3. Convert to Boolean")

choice = input("Enter your choice (1/2/3): ")

value = input("Enter any value: ")

if choice == "1":
    # Integer conversion
    try:
        print("Integer:", int(value))
    except:
        print("Cannot convert to Integer")

elif choice == "2":
    # Float conversion
    try:
        print("Float:", float(value))
    except:
        print("Cannot convert to Float")

elif choice == "3":
    # Boolean conversion (no error possible)
    print("Boolean:", bool(value))

else:
    print("Invalid Choice")

```

```python
# Write code to demonstrate type casting with list elements
value=["10","20","30.2","40.2"]
print("original list",value)
int_list=[]
for v in value:
  try:
    int_list.append(int(v))
  except:
    print("List not change in integer")
print("list change in integer",int_list)

float_list = []
for v in value:
    float_list.append(float(v))
print("Float List:", float_list)



```
